## Problem Definition

>Which players are good targets for recruitment? i.e. Which players have the biggest difference between their Value and Wage?

```python
# Create new subset dataframe
df = data_frame[['Name', 'Wage', 'Value']].copy()
df.head()
```

```python
# Convert Wage and Value from str to float
multipliers = {'K':1000, 'M':1000000, 'B':1000000000}

def string_to_float(string):
    if string[-1].isdigit():
        return float(string)
    mult = multipliers[string[-1]]
    return float(string[:-1]) * mult


df.Wage = df.Wage.replace('[\€]', '', regex=True).apply(string_to_float)
df.Value = df.Value.replace('[\€]', '', regex=True).apply(string_to_float)
df.head()
```

```python
# Create difference column
df['Difference'] = df['Value'] - df['Wage']
df.sort_values('Difference', ascending=False)
```

## Visualization

```python
sns.set()

graph = sns.scatterplot(x='Wage',
                        y='Value',
                        data=df).set(title='FIFA 2019 Player Value vs Wage')
```

```python
# Interactive visualization with bokeh
TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(Wage,Value)", "(@Wage, @Value)"),
    ("Name", "@Name",)
])

p = figure(title='FIFA 2019 Player Value vs Wage',
           x_axis_label='Wage',
           y_axis_label='Value',
           width=1200,
           height=800,
           tools=[TOOLTIPS])
p.circle('Wage', 'Value', size=10, source=df)
show(p)
```
